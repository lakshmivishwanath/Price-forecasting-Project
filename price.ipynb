```python
Q1 = df['avg_monthly_price'].quantile(0.25)
Q3 = df['avg_monthly_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['avg_monthly_price'] < lower_bound) |
              (df['avg_monthly_price'] > upper_bound)]

outlier_count = len(outliers)
total_count = len(df)

outlier_percentage = (outlier_count / total_count) * 100
outlier_percentage
```

# Predict NEXT MONTH price

```python
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestRegressor

# 1. Load dataset
df = pd.read_csv("price_data.csv")

# 2. Convert date
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')

# 3. Create time index
df['time_index'] = range(len(df))

# 4. Split features and target
X = df[['time_index']]
y = df['avg_monthly_price']   # If column name is different — tell me

# 5. Train model
model = RandomForestRegressor()
model.fit(X, y)

# 6. Save model as pickle file
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)

# 7. Print next-month forecast
next_time = [[len(df)]]
pred = model.predict(next_time)
print("Next Month Price Forecast =", pred[0])
```
